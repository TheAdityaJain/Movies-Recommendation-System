Feature Selection Part

```python
movies.columns
```

```python
movies.isnull().sum()
```

```python
movies=movies[['id','title','overview','genre']]
```

```python
movies
```

```python
movies['tags']=movies['overview']+movies['genre']
```

```python
movies
```

```python
new_data  = movies.drop(columns=['overview', 'genre']) #The drop() method in Pandas is used to remove columns (or rows) from a DataFrame. 
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=1000, stop_words='english') 
#CountVectorizer is used to convert a collection of text documents into a matrix of token counts
#Stop words are common words that are often filtered out during the preprocessing of text data
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=='The Godfather'].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("The Shawshank Redemption")
```
